```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Users\\abhijit.a.pande\\Machine Learning\\datasets\\fake_reg.csv")
```

```python
df.head()
```

```python
sb.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['feature1', 'feature2']].values
y = df['price'].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
mm.fit(x_train)
```

```python
x_train = mm.transform(x_train)
```

```python
x_test = mm.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 4, activation = "relu"))
model.add(Dense(units = 4, activation = "relu"))
model.add(Dense(units = 4, activation = "relu"))

model.add(Dense(1))
```

```python
model.compile(optimizer = 'Adam', loss = "mse")
```

```python
model.fit(x = x_train, y = y_train, epochs = 600, verbose = 2, workers = 4)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
fig = plt.figure()
```

```python
model.evaluate(x_test, y_test, verbose = 1)
```

```python
model.evaluate(x_train, y_train, verbose = 1)
```

```python
predictions = model.predict(x_test)
```

```python
pred_series = pd.Series(predictions.reshape(300,))
```

```python
real_df = pd.DataFrame(y_test, columns = ["real y"])
```

```python
pred_df = pd.concat([pred_series, real_df],axis = 1)
pred_df.columns=["Predicted","Real"]
```

```python
plt.scatter(pred_df["Predicted"], pred_df["Real"])
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(pred_df["Real"], pred_df["Predicted"])
```

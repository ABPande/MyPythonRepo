```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
x = np.linspace(0,5,11)
y = x ** 2
z = np.exp(x)
v = np.sqrt(x)
w = x ** 4
axes.plot(x,y, color = "green", label = "square")
axes.plot(x,z, color = "red", label = "exp")
axes.plot(x,v, color = "blue", label = "sqrt")
plt.legend(loc = 'best')
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (6,6))
toplot = [v,w,y,z]
colors = ["red","blue","green","cyan"]
variations = ["root", "quad", "square", "exp"]
i = 0
for axis in axes:
    for ax in axis:
        ax.plot(x, toplot[i], color = colors[i], linewidth = 2, linestyle = '--',alpha = 0.5,marker = "o", markersize = 10, markerfacecolor = 'yellow', markeredgecolor = 'black')
        ax.set_title("X variations " + variations[i])
        i = i + 1
        ax.set_xlim([0,6])
        ax.set_ylim([0,6])
fig.savefig("fig.png")
```

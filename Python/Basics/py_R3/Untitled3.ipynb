```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y,'r-')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("X vs Y")
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'r')
plt.subplot(1,3,2)
plt.plot(y,x,'b')
plt.subplot(1,3,3)
plt.plot(y,x ** .4,'g')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.4,0.42,0.42])
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes1.plot()
axes2.plot()
axes1
```

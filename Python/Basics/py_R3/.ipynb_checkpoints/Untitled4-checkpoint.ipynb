```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

fig  = plt.figure()
x = np.linspace(0,5,11)
print(x)
y = x ** 2
print(y)
ax = fig.add_axes([0.1,0.1,0.9,0.9])
print (ax)
ax.set_xlim([0,6])
ax.set_ylim([0,30])
ax.plot(x,y,color = '#AAFFC0',
        linewidth = 4, alpha = 0.99,
        linestyle = '--', marker = 'h',
        markersize = 20, 
        markerfacecolor = "yellow", 
        markeredgewidth = 2, 
        markeredgecolor = "green", label="x") 
```

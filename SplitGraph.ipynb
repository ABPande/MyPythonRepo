```python
import numpy as np
import pandas as pd

x = np.arange(0,101)
print(x)
```

```python
y = x * 2
```

```python
z = y ** 2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title ("Title")
ax.set_xlabel ("XLabel")
ax.set_ylabel ("YLabel")
ax.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,0.8])
ax2 = fig.add_axes([0.2,0.4,0.2,0.16]) 

ax1.plot(x,z)
ax2.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[0].width = 0.3
axes[1].plot(x,z)
```

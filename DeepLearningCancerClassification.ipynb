```python
x = df.drop('benign_0__mal_1', axis = 1)[:].values
```

```python
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
x_train = mm.fit_transform(x_train)
```

```python
x_test = mm.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
x_train.shape
```

```python
model = Sequential()
model.add(Dense(30,activation = "relu"))
model.add(Dense(15,activation = "relu"))
model.add(Dense(1,activation = "sigmoid"))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
model.fit(x = x_train, y = y_train, validation_data = (x_test, y_test), epochs=600)
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
model = Sequential()
model.add(Dense(30,activation = "relu"))
model.add(Dense(15,activation = "relu"))
model.add(Dense(1,activation = "sigmoid"))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
es = EarlyStopping(monitor = "val_loss", mode = "min", verbose = 1, patience = 10)
```

```python
model.fit(x = x_train, y = y_train, validation_data = (x_test, y_test), epochs=600, callbacks = [es])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
model = Sequential()
model.add(Dense(30,activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(15,activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(1,activation = "sigmoid"))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
es = EarlyStopping(monitor = "val_loss", patience = 10)
```

```python
model.fit(x_train, y_train, epochs = 600, validation_data = (x_test,y_test), callbacks = [es])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
predictions = model.predict_classes(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

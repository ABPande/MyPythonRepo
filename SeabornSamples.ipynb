```python
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
tips = sb.load_dataset('tips')
tips
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o', 'v'], scatter_kws = {'s':50})
```

```python
sb.lmplot(x = 'total_bill', y = 'tip',
          data = tips, 
          row = 'smoker', col = 'sex', hue = 'day',
          markers = ['o', 'v','v','v'])
```

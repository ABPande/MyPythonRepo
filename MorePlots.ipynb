```python
import seaborn as sb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sb.load_dataset("tips")
flights = sb.load_dataset("flights")
```

```python
sb.heatmap(flights.corr())
plt.tight_layout()
```

```python
sb.heatmap(tips.corr(), annot = True, cmap = "magma")
plt.tight_layout()
```

```python
fp = flights.pivot_table(index = "month", columns = "year", values = "passengers")
sb.heatmap(fp, linecolor = "white", linewidth = "1")
```

```python
sb.clustermap(fp, cmap = "coolwarm", standard_scale = 1)
```

```python
iris = sb.load_dataset('iris')
```

```python
iris.head()
```

```python
sb.pairplot(iris)
```

```python
pg = sb.PairGrid(iris)
pg.map_lower(sb.kdeplot)
pg.map_upper(sb.regplot)
pg.map_diag(sb.distplot)
```

```python
tips = sb.load_dataset("tips")
pg = sb.FacetGrid(data = tips, col = "time", row = "smoker")
pg.map(plt.scatter, "total_bill", "tip")
```

```python
sb.lmplot(x = 'total_bill', y ='tip', hue = "sex", data = tips, scatter_kws = {'s':100}, markers = ["v","o"], col = "sex", row = "time", aspect = 1)
```

```python
plt.figure(figsize = (10,7))
sb.set_context("poster", font_scale=1)
sb.set_style("ticks")
sb.countplot(tips["sex"])
sb.despine()
```

```python
sb.lmplot(x = "total_bill", y = "tip", data = tips, hue = "sex", palette = "seismic")
```

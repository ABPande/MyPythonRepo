```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from plotly import __version__
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode, plot, iplot
init_notebook_mode(connected = True)
cf.go_offline()
df1 = pd.DataFrame(np.random.randn(100,4), columns = ['a','b','c','d'])
df2 = pd.DataFrame({'Category': ['A','B','C'] ,
                   'Values': [32,43,50]})
%matplotlib inline
```

```python
df1.iplot()
```

```python
df1.iplot(kind = "scatter", x = 'a', y = 'b', mode = 'markers')
```

```python
df2.iplot(kind = "bar", x = 'Category' , y = 'Values')
```

```python
df1.sum().iplot(kind = "bar")
```

```python
df1.iplot(kind = "box")
```

```python
df3 = pd.DataFrame ({'x': [1,2,3,4,5], 'y': [10,20,30,40,50],
                    'z': [500,400,300,200,100]} )
```

```python
df3.iplot(kind = "surface", colorscale = "rdylbu")
```

```python
import seaborn as sb
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
iris = sb.load_dataset('iris')

sb.pairplot(iris)
```

```python
g = sb.PairGrid(iris)
g.map_diag(sb.distplot)
g.map_lower(plt.scatter)
g.map_upper(sb.kdeplot)
```

```python
tips = sb.load_dataset('tips')
```

```python
g = sb.FacetGrid(data = tips, col = 'time', row = 'smoker', hue = 'sex')
g.map(sb.distplot,'total_bill')
g.map(plt.scatter,'total_bill', 'tip')
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', markers = ['o', 'v'], scatter_kws = {'s':50})
```

```python
sb.lmplot(data = tips, x = 'tip', y = 'total_bill', col = 'sex', row = 'time', hue = 'day', aspect = 1, size = 8)
```

```python
plt.figure (figsize = (12,6))
sb.set_style('ticks')
sb.set_context('poster', font_scale = 1)
sb.countplot ('sex', data = tips)
sb.despine (left = True, bottom = True)
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', markers = ['o', 'v'], scatter_kws = {'s':50}, palette = 'seismic')
```

```python
titanic = sb.load_dataset('titanic')
titanic
```

```python
sb.jointplot(x = 'fare', y = 'age', data = titanic, kind = 'scatter' )
```

```python
sb.distplot(titanic['fare'], kde = False, color = 'red', bins = 25)
```

```python
sb.boxplot(x = 'class', y = 'age', data = titanic)
```

```python
sb.swarmplot(x = 'class', y = 'age', data = titanic)
```

```python
sb.countplot(x = 'sex', data = titanic)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
tips = sb.load_dataset("tips")
```

```python
sb.barplot(tips["sex"],tips["total_bill"], hue = 'smoker', data = tips, estimator = np.mean)
```

```python
sb.countplot(tips['smoker'])
```

```python
sb.boxplot(tips["day"], tips['total_bill'],  hue = tips["smoker"])
```

```python
sb.violinplot(x = "day", y = "total_bill", hue = "smoker", data = tips, split = True)
```

```python
sb.stripplot(x= "day", y = "total_bill", data = tips, jitter = True)
```

```python
sb.swarmplot(x = "day", y = "total_bill", data = tips)
```

```python
sb.factorplot(x = "day", y = "total_bill", data = tips, kind = "bar")
```

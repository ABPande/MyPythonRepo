```python
import matplotlib.pyplot as plt
%matplotlib inline
#plt.show() if you arent using jupyter
import numpy as np
x = np.linspace(0, 5, 11)
y = np.sqrt(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, label = "root")
plt.plot(x, x ** 2, label = "square")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,label = 'root')
plt.subplot(1,2,2)
plt.plot(x,x**2, label = 'sq', color = 'red')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,"green")
axes.plot(x,x**2, color = "red")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.5,0.2,0.4,0.4])
axes1.plot(x,y,color = "red")
axes2.plot(x,x**2,color = "green")
```

```python
x  = df.drop("zipcode", axis = 1)
x = x.drop("date", axis = 1)
x = x.drop("price", axis = 1)
x = x.drop("lat", axis = 1)
x = x.drop("long", axis = 1)
x = x.drop("month", axis = 1)
```

```python
y = df["price"].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
mm.fit(x_train)
```

```python
x_train = mm.transform(x_train)
```

```python
x_test = mm.transform(x_test)
```

```python
x_test
```

```python
x
```

```python
model = Sequential()
model.add(Dense(18, activation = "relu"))
model.add(Dense(18, activation = "relu"))
model.add(Dense(18, activation = "relu"))

model.add(Dense(1))

model.compile(optimizer="adam", loss = "mae")
```

```python
model.fit(x = x_train, y = y_train, epochs = 100, verbose = 2, validation_data=(x_test,y_test))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
predictions = model.predict(x_test)
```

```python
predictions.reshape(6415,1)
```

```python
pred_df = pd.DataFrame(pd.Series(predictions.reshape(6415,)))
y_df = pd.DataFrame(pd.Series(y_test.reshape(6415,)))
outcome = pd.concat([pred_df, y_df],axis = 1)
```

```python
outcome.columns = ["Predicted", "Real"] 
```

```python
outcome.head()
```

```python
mean_absolute_error(outcome["Real"], outcome["Predicted"])
```

```python
mean_squared_error(outcome["Real"], outcome["Predicted"])
```

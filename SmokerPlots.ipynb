```python
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
tips = sb.load_dataset('tips')
flights = sb.load_dataset ('flights')
sb.boxplot(x = 'day' , y = 'tip', data = tips, hue = 'smoker' )
```

```python
iris = sb.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sb.pairplot(iris)
```

```python
irispg = sb.PairGrid(iris)
irispg.map (plt.scatter)
irispg.map_diag(sb.distplot)
irispg.map_lower(sb.kdeplot)
irispg.map_upper(plt.scatter)
```

```python
sb.distplot(iris['sepal_length'])
```

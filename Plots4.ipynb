```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.show() #if you dont use inline
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x ** 2
```

```python
plt.plot(x,y,'r-',marker = "o")
```

```python
plt.xlabel('Number')
plt.ylabel("Squared")
plt.title("Exponential: Power 2")
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
```

```python
fig, axes = plt.subplots(3,3)
plt.tight_layout()
for current_axes_row in axes:
    for current_element in current_axes_row:
        current_element.plot(x,y)
```

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(2,2,figsize = (8,6))
for row_ in axes:
    for ele_ in row_:
        ele_.set_title("plot")
        ele_.set_xlabel("xlabel")
        ele_.set_ylabel("ylabel")
        ele_.plot(x,y,label = "Squared")
        ele_.legend(loc = 0)
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xlim([0,5])
ax.set_ylim([0,30])
ax.plot(x,y,color = "green",linestyle="--",linewidth = "3",marker = "o", markersize = "15", markerfacecolor = "red", markeredgecolor = "red")
```

```python
import seaborn as sb
%matplotlib inline
tips = sb.load_dataset('tips')
tips.head()
```

```python
sb.lmplot(x = 'total_bill', 
          y = 'tip', 
          data = tips, 
          hue = 'sex', 
          markers = ['o', 'v'], 
          scatter_kws={'s':50} )
```

```python
sb.lmplot(x = 'total_bill', y = 'tip', data = tips, col = 'sex', row = 'day', aspect = 0.6, height = 8)
```
